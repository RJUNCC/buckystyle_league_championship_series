```python
query = f"""
SELECT 
    b.discord_username,
    a.season_id,
    b.discord_id,
    a.games_played,
    a.wins,
    a.losses,
    a.avg_score,
    a.goals_per_game,
    a.assists_per_game,
    a.saves_per_game,
    a.shots_per_game,
    a.shot_percentage
FROM blcs_player_statistics a
INNER JOIN blcs_player_mappings b
ON a.player_id=b.ballchasing_player_id
"""
df = pd.read_sql(sql=query, con=engine)
df["player_name"] = df["discord_username"].str.extract(r"^([^|]+)")
df["team_name"] = df["discord_username"].str.extract(r"\|\s*(.+)$")
df
```
