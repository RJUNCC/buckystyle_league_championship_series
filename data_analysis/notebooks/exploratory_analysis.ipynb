```python
def fetch_previous_team_stats(group_id, token=config._ballchasing_token):

    url = f"https://ballchasing.com/api/groups/{group_id}"
    headers = {"Authorization": token}
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    data = response.json()
    df = pd.json_normalize(data, record_path=["teams"])
    return df

team_df = fetch_previous_team_stats(group_id)
```

```python
team_df = team_df.sort_values(by="name").reset_index(drop=True)
team_df
```

```python
team_df.loc[3, "cumulative.games"] = team_df.loc[3, "cumulative.games"] + team_df.loc[4, "cumulative.games"]
team_df.loc[3, "cumulative.core.goals"] = team_df.loc[3, "cumulative.core.goals"] + team_df.loc[4, "cumulative.core.goals"]
team_df.loc[3, "cumulative.core.goals_against"] = team_df.loc[3, "cumulative.core.goals_against"] + team_df.loc[4, "cumulative.core.goals_against"]
team_df.loc[3, "cumulative.demo.inflicted"] = team_df.loc[3, "cumulative.demo.inflicted"] + team_df.loc[4, "cumulative.demo.inflicted"]
team_df.loc[3, "cumulative.demo.taken"] = team_df.loc[3, "cumulative.demo.taken"] + team_df.loc[4, "cumulative.demo.taken"]
team_df.loc[3, "cumulative.win_percentage"] = (team_df.loc[3, "cumulative.wins"] + 1) / team_df.loc[3, "cumulative.games"] *100
team_df.loc[3, "cumulative.core.shots"] = team_df.loc[3, "cumulative.core.shots"] + team_df.loc[4, "cumulative.core.shots"]
team_df.loc[3, "cumulative.core.shots_against"] = team_df.loc[3, "cumulative.core.shots_against"] + team_df.loc[4, "cumulative.core.shots_against"]
```

```python
team_df = team_df.drop(4)
team_df
```

```python
# Calculate Z-scores with weights
team_df["Goal Diff"] = team_df["cumulative.core.goals"] - team_df["cumulative.core.goals_against"]
team_df["Demo Diff"] = team_df["cumulative.demo.inflicted"] - team_df["cumulative.demo.taken"]
team_df["Shots Diff"] = team_df["cumulative.core.shots"] - team_df["cumulative.core.shots_against"]
team_df["Win % Zscore"] = zscore(team_df["cumulative.win_percentage"]) * config.win_perc_weight
team_df["Goal Diff Zscore"] = zscore(team_df["Goal Diff"]) * config.goal_diff_weight
team_df["Demo Diff Zscore"] = zscore(team_df["Demo Diff"]) * config.demo_diff_weight
team_df["Shots Diff Zscore"] = zscore(team_df["Shots Diff"]) * config.shot_diff_weight
# team_df["Team Played Win % Zscore"] = zscore(team_df["Team Played Win %"]) * config.strength_of_schedule
```

```python
# Calculate EPI Score
team_df["EPI Score"] = team_df[[
    "Win % Zscore", 
    "Goal Diff Zscore", 
    "Demo Diff Zscore", 
    "Shots Diff Zscore", 
    # "Team Played Win % Zscore"
]].sum(axis=1)
```

```python
features_of_interest = ["name", "cumulative.win_percentage", "game_average.core.goals", "game_average.core.goals_against", "cumulative.core.goals", 
                        "cumulative.core.goals_against", "game_average.core.shots", "game_average.core.shots_against", "cumulative.core.shots", "cumulative.core.shots_against", "EPI Score"
                        ]
```

```python
team_df = team_df.drop(columns=["cumulative.core.goals", "cumulative.core.goals_against", "cumulative.core.shots", "cumulative.core.shots_against"], axis=1)
```
